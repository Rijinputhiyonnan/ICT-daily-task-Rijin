```python



condition = (df['poutcome'] == 'nonexistent') | (
    (df['job'] == 'unknown') |
    (df['marital'] == 'unknown') |
    (df['education'] == 'unknown') |
    (df['default'] == 'unknown') |
    (df['housing'] == 'unknown') |
    (df['loan'] == 'unknown')
)

index_numbers = df[condition].index.tolist()
index_numbers = df[condition].index.tolist()
condition.sum()
print(f"Index numbers of rows with 'nonexistent' in 'poutcome' or 'unknown' in any specified column: {index_numbers}")
print(condition.sum())
```

```python

columns_to_check = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'poutcome']

# Create a boolean DataFrame where True indicates 'unknown' or 'nonexistent'
boolean_df = df[columns_to_check].apply(lambda x: x.isin(['unknown', 'nonexistent']))

# Sum the boolean values across the specified columns for each row
row_sums = boolean_df.sum(axis=1)

# Find the indices where the sum is 2 or more and y == 0
condition = (row_sums >= 2) & (df['y'] == 0)

# Get the total count of such rows
total_count = condition.sum()

print(f"Total number of rows with two or more 'unknown' or 'nonexistent' values and y == 0: {total_count}")

```

```python
df_cleaned_1 = df[~condition]
df_cleaned_1['y'].value_counts()
```

Hyper parameter tuning of SVM model


```python
from sklearn.model_selection import GridSearchCV

from sklearn.metrics import  make_scorer
```

```python

# Define the SVC model
svc_model = SVC()

# Define the hyperparameter grid
param_grid = {
    'C': [0.1, 1, 10, 100],
    'degree': [2, 3, 4, 5],
    'coef0': [0.0, 0.5, 1.0],
    'kernel': ['poly']
}

# Define the scoring method
f1_scorer = make_scorer(f1_score, average='weighted')

# Use a different backend to avoid the os.fork() issue
from joblib import parallel_backend

with parallel_backend('threading'):
    # Initialize GridSearchCV with n_jobs=-1 for parallel processing
    grid_search = GridSearchCV(svc_model, param_grid, scoring=f1_scorer, cv=5, n_jobs=-1)

    # Fit the model to the training data
    grid_search.fit(x_train, y_train)

# Get the best model from grid search
best_model = grid_search.best_estimator_

# Make predictions with the best model
y_pred = best_model.predict(x_test)

# Calculate and print the F1 score
f1 = f1_score(y_test, y_pred, average='weighted')
print(f"Best Hyperparameters: {grid_search.best_params_}")
print(f"F1 Score: {f1:.4f}")

# Extract the results from the grid search
results = grid_search.cv_results_

# Convert to a DataFrame for easier plotting
df = pd.DataFrame(results)

# Plot F1 score against the degree of the polynomial kernel
sns.lineplot(data=df, x='param_degree', y='mean_test_score', marker='o')
plt.title('F1 Score vs. Polynomial Degree')
plt.xlabel('Degree')
plt.ylabel('Mean F1 Score')
plt.grid(True)
plt.show()

# Plot F1 score against the C parameter
sns.lineplot(data=df, x='param_C', y='mean_test_score', marker='o')
plt.title('F1 Score vs. C Parameter')
plt.xlabel('C')
plt.ylabel('Mean F1 Score')
plt.xscale('log')  # Use log scale for better visualization
plt.grid(True)
plt.show()

# Plot F1 score against coef0
sns.lineplot(data=df, x='param_coef0', y='mean_test_score', marker='o')
plt.title('F1 Score vs. coef0')
plt.xlabel('coef0')
plt.ylabel('Mean F1 Score')
plt.grid(True)
plt.show()
```
