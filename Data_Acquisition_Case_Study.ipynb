```python
Q1=death_rate['ESTIMATE'].quantile(.25)
Q3=death_rate['ESTIMATE'].quantile(.75)

IQR=Q3-Q1
lower=Q1-1.5*IQR
upper=Q3+1.5*IQR

Outlier=death_rate[(death_rate['ESTIMATE']<lower)] | death_rate[(death_rate['ESTIMATE']>upper)]
Outlier
```
